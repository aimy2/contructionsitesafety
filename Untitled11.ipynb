```python
import cv2
import os

# Path to your video
video_path = "/content/construction.mp4"

# Folder to save frames
output_folder = "/content/frames"
os.makedirs(output_folder, exist_ok=True)

# Open video
cap = cv2.VideoCapture(video_path)

frame_count = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Exit when video ends

    # Save frame as image
    frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.jpg")
    cv2.imwrite(frame_filename, frame)

    frame_count += 1

cap.release()
cv2.destroyAllWindows()

print(f"✅ Done! Extracted {frame_count} frames to '{output_folder}'")
```

```python
from ultralytics import YOLO
import cv2
import os

# Load YOLO model
model = YOLO("/content/runs/detect/train3/weights/best.pt")  # your trained model path

# Required safety items (labels exactly as in your dataset)
required_items =['glasses', 'gloves', 'helmet', 'shoes', 'worker']
# Open video
cap = cv2.VideoCapture("/content/construction.mp4")
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# Save the output video to the default content directory
output_path = "/content/output_labeled.mp4"
out = cv2.VideoWriter(output_path, fourcc, 30.0,
                      (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame)
    detections = results[0].names
    detected_items = set()

    # Draw detections
    for box in results[0].boxes:
        cls_id = int(box.cls[0])
        label = results[0].names[cls_id]
        detected_items.add(label)

        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.6, (0, 255, 0), 2)

    # Check missing safety items
    missing_items = [item for item in required_items if item not in detected_items]
    if missing_items:
        alert_text = f"ALERT! Missing: {', '.join(missing_items)}"
        cv2.putText(frame, alert_text, (50, 50), cv2.FONT_HERSHEY_SIMPLEX,
                    1.0, (0, 0, 255), 3)
        print(alert_text)  # You can replace with buzzer/email/etc.

    # Save frame to output video
    out.write(frame)
    # cv2.imshow("Safety Detection", frame) # Disable imshow in Colab

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"✅ Done! Processed video saved to '{output_path}'")
```
